```python
import pandas            as pd
import numpy             as np
import matplotlib.pyplot as plt
import seaborn           as sns
```

```python
data_dir  = 'data/youtube'
data_file = 'videos-stats.csv'

df = pd.read_csv(f'{data_dir}/{data_file}')
df.info()
```

### Analisando valores vazios

```python
df[ df['Likes'].isna() ]
```

```python
df.dropna(inplace=True)
```

```python
df[['Likes', 'Comments']].info()
```

```python
sns.scatterplot(data=df, x='Likes', y='Comments')
```

```python
likes_df = df[ df['Likes'] <= df['Likes'].median() ]
likes_df.info()
```

```python
x_df = likes_df['Likes']
y_df = likes_df['Comments']
sns.scatterplot(x=x_df, y=y_df)
```

```python
ready_df = likes_df[ likes_df['Comments'] <= likes_df['Comments'].median() ]
ready_df.info()
```

```python
sns.scatterplot(x=ready_df['Likes'], y=ready_df['Comments'])
```
